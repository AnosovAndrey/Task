```python
#Normalisation module

X = np.array(data)[:,2]
X = np.reshape(X,(-1,1600))

print("Shape of X",X.shape)
print("Examples of X\n",X[:1])

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
print("Examples of X_scaled\n",X_scaled[:1])

# X_normalized = normalize(X, axis=0)
# print("Examples of X_normalised\n",X_normalized[:3])

#inverse norm
# X_inv = scaler.inverse_transform(X_scaled)
# print("Examples of X_inv\n",X_inv[:3])

Y = np.array(otrezok)

print("Shape of Y",Y.shape)
print("Examples of Y\n",Y[:1])

y_scaler = MinMaxScaler()
Y_scaled = y_scaler.fit_transform(Y)
print("Examples of Y_scaled\n",Y_scaled[:1])

# Y_inv = y_scaler.inverse_transform(Y_scaled)
# print("Examples of Y_inv\n",Y_inv[:3])
```

```python
total_length=len(X)
train_length=int(0.8*total_length)
test_length=int(0.2*total_length)
# print(train_length)
# print(test_length)

X_train=X_scaled[:train_length]
X_test=X_scaled[train_length:]
y_train=Y_scaled[:train_length]
y_test=Y_scaled[train_length:]

print("Length of train set x:",X_train.shape[0],"y:",y_train.shape[0])
print("Length of test set x:",X_test.shape[0],"y:",y_test.shape[0])

#print("Shape of y_train",X_train.shape)
```

```python
#Neural network module
from keras.models import Sequential 
from keras.layers import Dense,Activation,Dropout 
from keras.layers.normalization import BatchNormalization 
from keras.utils import np_utils
```

```python
model=Sequential()

# model.add(Dense(1600,input_dim=1600,activation='relu'))
# model.add(Dense(2000,activation='relu'))
# model.add(Dense(2000,activation='relu'))
# model.add(Dense(2000,activation='relu'))
# #model.add(Dense(300,activation='relu'))
# model.add(Dropout(0.3))
# model.add(Dense(7,activation='sigmoid'))
# model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])


#### 0.5
# model.add(Dense(2000,input_dim=1600,activation='sigmoid')) 
# model.add(Dense(2500,activation='sigmoid'))
# model.add(Dense(300,activation='sigmoid'))
# model.add(Dropout(0.3))
# model.add(Dense(7,activation='sigmoid'))
# model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

##### 0.57
# model.add(Dense(1000,input_dim=1600,activation='relu'))
# model.add(Dense(800,activation='relu'))
# model.add(Dense(600,activation='relu'))
# model.add(Dropout(0.3))
# model.add(Dense(7,activation='sigmoid'))
# model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.add(Dense(16000,input_dim=1600,activation='relu'))
model.add(Dense(7000,activation='relu'))
model.add(Dense(700,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(7,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
#model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy']) ### 0.4


#model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
# model.compile(
#     optimizer=keras.optimizers.RMSprop(1e-3),
#     loss=['binary_crossentropy'], metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=25,epochs=100,verbose=1)
```

```python
prediction=model.predict(X_test)
length=len(prediction)
y_label=np.argmax(y_test,axis=1)
predict_label=np.argmax(prediction,axis=1)

accuracy=np.sum(y_label==predict_label)/length * 100 
print("Accuracy of the dataset",accuracy )
```
