```python
X = np.array(data)[:,2]
X = np.reshape(X,(-1,6400))
print("Examples of X\n",X[:1])

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
print("Examples of X_scaled\n",X_scaled[:1])

Y = np.array(otrezok)

print("Shape of Y",Y.shape)
print("Examples of Y\n",Y[:1])

y_scaler = MinMaxScaler()
Y_scaled = y_scaler.fit_transform(Y)
print("Examples of Y_scaled\n",Y_scaled[:1])
```

```python
total_length=len(X)
train_length=int(0.8*total_length)
test_length=int(0.2*total_length)
print(train_length)
print(test_length)
```

```python
#Neural network module
from keras.models import Sequential 
from keras.layers import Dense,Activation,Dropout,LSTM,Embedding 
from keras.layers.normalization import BatchNormalization 
from keras.utils import np_utils
```

```python
X_input = np.reshape(X_scaled,(-1, 80, 80)) #np.reshape(X, X.shape + (1,))
# print(X_input.shape)
# print(X_input[:1])
Y_output1 = np.array(Y_scaled)[0:,0:6]
Y_output2 = np.array(Y_scaled)[0:,6:7]

X_train2=X_input[:train_length]
X_test2=X_input[train_length:]
y_train21=Y_output1[:train_length]
y_test21=Y_output1[train_length:]
y_train22=Y_output2[:train_length]
y_test22=Y_output2[train_length:]

print(X_train2.shape)
```

```python
from keras.layers import Input, Conv2D, MaxPooling2D, GlobalAveragePooling2D
from keras.models import Model
```

```python
def get_model():
    inputs = Input(shape=(80,80,1))
    x = Conv2D(32, (3,3), activation='relu')(inputs)
    x = MaxPooling2D((3,3))(x)
    x = Conv2D(32, (3,3), activation='relu')(x)
    x = MaxPooling2D((3,3))(x)
    x = Conv2D(64, (3,3), activation='relu')(x)
    x = GlobalAveragePooling2D()(x)
    classifier_head = Dropout(0.3)(x)
    # Notice the name of the layer.
    classifier_head = Dense(1, activation='sigmoid', name='weight')(classifier_head)
    reg_head = Dense(64, activation='relu')(x)
    reg_head = Dense(32, activation='relu')(reg_head)
    # Notice the name of the layer.
    reg_head = Dense(6, activation='sigmoid', name='coordinates')(reg_head)
    return Model(inputs=[inputs], outputs=[classifier_head, reg_head])
```

```python
# losses = {'weight': 'binary_crossentropy',
#           'coordinates': 'mse'}
losses = {'weight': 'mse',
          'coordinates': 'mse'}
loss_weights = {'weight': 1.0,
                'coordinates': 1.0}
metrics = { 'weight': 'mae', #'weight': 'mean_absolute_error',
                'coordinates': 'acc'}
```

```python
x_model = get_model()
x_model.summary()
x_model.compile('adam', loss=losses, loss_weights=loss_weights, metrics=metrics)
```

```python
H = x_model.fit(X_train2, {'coordinates': y_train21, 'weight': y_train22},validation_data=(X_test2,{'coordinates': y_test21, 'weight': y_test22}),batch_size=25,epochs=200,verbose=1)
```

```python
result_2=x_model.predict(X_test2)
#print(result_2)
res = np.column_stack((result_2[1], result_2[0]))
#print(res[:1])
result_2_inv = y_scaler.inverse_transform(res)
#print(result_2_inv)
```
